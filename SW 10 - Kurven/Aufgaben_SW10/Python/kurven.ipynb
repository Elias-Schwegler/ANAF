```python
# Bibliotheken

import numpy as np
import matplotlib.pyplot as pl
import ipywidgets as iw

%matplotlib widget
```

```python
# Fig

def Fig ():

    fig = pl.figure (figsize = (9, 5))

    cnv = fig.canvas
    cnv.toolbar_visible = False
    cnv.header_visible = False
    cnv.footer_visible = False

    return fig
```

## Beispiele ebener Kurven

### Kurven interaktiv darstellen

```python
# InteractCurve (curve, a, b)

def InteractCurve (curve, a, b):

    tt = np.linspace (a, b, 200)
    pp = curve (tt)

    Fig()
    pl.axis ('equal')
    pl.plot (pp [0], pp [1], color = 'blue', zorder = 2)
    pnt, = pl.plot ([0], [0], 'o', color = 'red', zorder = 4)
    pl.show()

    def Callback (t0):
        p = curve (np.array ([t0]))
        pnt.set_data (p [0], p [1])

    lay = iw.Layout (width = '900px')
    sl = iw.FloatSlider (min = a, max = b, step = (b - a) / 1000, value = a, layout = lay)
    iw.interact (Callback, t0 = sl)
```

### Beispiel 1: <code> Alpha </code>

$\displaystyle
    c_1 \, (t)
        \; = \;
        \begin{pmatrix}
            t^2 - 1 \\
            t - t^3
        \end{pmatrix}
        \, \mbox{,} \qquad
    t \, \in \left[ \, - \frac{5}{4} , \frac{5}{4} \, \right]

```python
def Alpha (t):
    x = t ** 2 - 1
    y = t - t ** 3
    return np.array ([x, y])

InteractCurve (Alpha, -1.25, 1.25)
```

## Ableitung von Kurven, Tangentialvektor

### Mathematische Grundlagen

$\begin{align*}
    \dot{c} \, (t)
        & \; = \; \lim_{t \, \to \, 0} \, \frac{c \, (t + h) - c \, (t)}{h}
        \; = \; \lim_{t \, \to \, 0} \, \frac{1}{h} \, \left[ \,  c \, (t + h) - c \, (t) \, \right]
        \; = \; \lim_{t \, \to \, 0} \, \frac{1}{h} \, \left[ \,
            \begin{pmatrix} x \, (t + h) \\ y \, (t + h) \end{pmatrix}
            - \begin{pmatrix} x \, (t) \\ y \, (t) \end{pmatrix} \, \right]
        \; = \; \lim_{t \, \to \, 0} \, \frac{1}{h}
            \begin{pmatrix} x \, (t + h) - x \, (t) \\ y \, (t + h) - y \, (t) \end{pmatrix}
        \\[6pt]
        & \; = \; \lim_{t \, \to \, 0}
            \begin{pmatrix} 
                \displaystyle \frac{x \, (t + h) - x \, (t)}{h} \\ 
                \displaystyle \frac{y \, (t + h) - y \, (t)}{h} \end{pmatrix}
        \; = \;
            \begin{pmatrix} 
                \displaystyle \lim_{t \, \to \, 0} \, \frac{x \, (t + h) - x \, (t)}{h} \\ 
                \displaystyle \lim_{t \, \to \, 0} \, \frac{y \, (t + h) - y \, (t)}{h} \end{pmatrix}
        \; = \;
            \begin{pmatrix} \dot{x} \, (t) \\ \dot{y} \, (t) \end{pmatrix}
\end{align*}$

**Merke: Kurven werden komponentenweise differenziert!**

### Tangenten (symbolisch) interaktiv darstellen

```python
# PlotVector

def PlotVector (p, v, **kwargs):
    wd = 0.003
    return pl.quiver (p [0], p [1], v [0], v [1], **kwargs,
        angles = 'xy', scale = 1, scale_units = 'xy', width = wd, headwidth = 6, headlength = 10, headaxislength = 9)
```

```python
# InteractTangentSym (curve, tangent, a, b)

def InteractTangentSym (curve, tangent, a, b):

    tt = np.linspace (a, b, 200)
    pp = curve (tt)
    qq = pp + tangent (tt)

    Fig()
    pl.axis ('equal')
    pl.plot (pp [0], pp [1], color = 'blue', zorder = 2)
    pl.plot (qq [0], qq [1], 'red', linewidth = 0)

    t0 = a
    p = curve (t0)
    v = tangent (t0)
    pnt, = pl.plot (p [0], p [1], 'o', color = 'red', zorder = 4) 
    tng = PlotVector (p, v, color = 'red', zorder = 4)
    pl.show()

    def Callback (t0):

        t = np.array ([t0])
        p = curve (t)
        v = tangent (t)

        pnt.set_data (p [0], p [1])
        tng.set_offsets ([p [0] [0], p [1] [0]])
        tng.set_UVC (v [0] [0], v [1] [0])

    lay = iw.Layout (width = '900px')
    sl = iw.FloatSlider (min = a, max = b, step = (b - a) / 1000, value = a, layout = lay)
    iw.interact (Callback, t0 = sl)
```

### Beispiel 1a: <code> Alpha </code>, Originalgeschwindigkeit

$\displaystyle
    c_1 \, (t)
        \; = \;
        \begin{pmatrix}
            t^2 - 1 \\
            t - t^3
        \end{pmatrix}
        \,\mbox{,} \qquad
    \dot{c}_1 \, (t)
        \; = \;
        \begin{pmatrix}
            2 \, t \\
            1 - 3 \, t^2
        \end{pmatrix}
        \,\mbox{,} \qquad
    t \, \in \left[ \, - \frac{5}{4}, \, \frac{5}{4} \right]
$

```python
def AlphaTangent (t):
    x = 2 * t
    y = 1 - 3 * t ** 2
    return np.array ([x, y])
```

```python
InteractTangentSym (Alpha, AlphaTangent, -1.2, 1.2)
```

### Beispiel 1b: <code> Alpha </code>, Geringere Geschwindigkeit

$\displaystyle
    c_5 \, (t)
        \; = \;
        c_1 \, \left( \frac{t}{4} \right)
        \; = \;
        \begin{pmatrix}
            \displaystyle \left( \frac{t}{4} \right) ^ 2 \\[3pt]
            \displaystyle \frac{t}{4} - \left( \frac{t}{4} \right) ^ 3
        \end{pmatrix}
        \; = \;
        \begin{pmatrix}
            \displaystyle \frac{t^2}{16} \\[3pt]
            \displaystyle \frac{t}{4} - \frac{t^3}{64}
        \end{pmatrix}
        \,\mbox{,} \qquad
    t \, \in \left[ \, -5, 5 \, \right]
$

$\displaystyle
    \dot{c}_5 \, (t)
        \; = \;
        \begin{pmatrix}
            \displaystyle \frac{t}{8} \\[3pt]
            \displaystyle \frac{1}{4} - \frac{3 \, t^2}{64}
        \end{pmatrix}
        \; = \;
        \begin{pmatrix}
            \displaystyle \frac{1}{2} \cdot \frac{t}{4} \\[3pt]
            \displaystyle \frac{1}{4} - 3 \cdot \left( \frac{t}{4} \right) ^ 2
        \end{pmatrix}
        \; = \;
        \frac{\displaystyle 1}{\displaystyle 4} \,
        \begin{pmatrix}
            \displaystyle 2 \cdot \frac{t}{4} \\[3pt]
            \displaystyle 1 - \frac{3}{4} \cdot \left( \frac{t}{4} \right) ^ 2
        \end{pmatrix}
        \; = \;
        \frac{1}{4} \, \cdot \, \dot{c}_1 \, \left( \frac{t}{4} \right)
        \qquad\textbf{Kettenregel!}
$

```python
InteractTangentSym (lambda t: Alpha (t / 4), lambda t: AlphaTangent (t / 4) / 4, -5, 5)
```

### Kurven umparametrisieren

$\begin{align}
    c: \quad \mathbb{R} \, \ni \, [ \, u_0, u_1 \, ]
        & \; \longrightarrow \; \mathbb{R}^n
        & \qquad
    u: \quad \mathbb{R} \, \ni \, [ \, t_0, t_1 \, ]
        & \; \longrightarrow \; [ \, u_0, u_1 \, ] \, \in \, \mathbb{R}
        & \qquad
    c \, \circ \, u: \quad \mathbb{R} \, \ni \, [ \, t_0, t_1 \, ]
        & \; \longrightarrow \; \mathbb{R}^n
        \\[6pt]
    u
        & \; \longmapsto \; c \, (u)
        & \qquad
    t
        & \; \longmapsto \; u \, (t)
        & \qquad
    t
        & \; \longmapsto \; c \, (u \, (t))
\end{align}$

$\begin{align}
    \left. \frac{\mathrm{d}}{\mathrm{d} t} \, c \, (u \, (t)) \right\vert _ {t \, = \, t_0}
        & \; = \; \frac{\mathrm{d c}}{\mathrm{d} u} \, (u \, (t_0))
            \cdot \frac{\mathrm{d u}}{\mathrm{d} t} \, (t_0))
\end{align}$

### Beispiel 1c: <code> Alpha </code>, Beschleunigen und Bremsen

$\begin{align*}
    u: [ \, 0, 10 \, ] 
        & \; \longrightarrow \; \left[ \, - \frac{5}{4}, \frac{5}{4} \right]
    & \qquad u \, (0)
        & \; = \; - \frac{5}{4}
    & \qquad u \, (10)
        & \; = \; \frac{5}{4}
    & \qquad u \, (t)
        & \; = \; \frac{- t^3 + 15 \, t - 250}{200}
        \\[6pt]
    t 
        & \; \longmapsto \; u \, (t)
    & u' \, (0)
        & \; = \; 0
    & u' \, (10)
        & \; = \; 0
    & u' \, (t)
        & \; = \; \frac{-3 \, t^2 + 30 \, t}{200}
\end{align*}$

```python
def u (t):
    return (- t ** 3 + 15 * t ** 2 - 250) / 200
```

```python
def du (t):
    return (-3 * t ** 2 + 30 * t) / 200
```

```python
InteractTangentSym (lambda t: Alpha (u (t)), lambda t: AlphaTangent (u (t)) * du (t), 0, 10)
```

### Numerisch differenzieren

$\displaystyle
    \dot{c} \, (t)
        \; = \; \lim_{h \, \to \, 0} \, \frac{\displaystyle c \, (t + h) - c \, (t)}{\displaystyle h}
        \; \approx \; \frac{\displaystyle c \, (t + h) - c \, (t)}{\displaystyle h}
       \quad\text{f\"ur kleine $h$, bei 16 Stellen Rechengenauigkeit $h_{\text{opt}}=10^{-8}$} 
$

```python
def NumDiff (c, t):
    h = 1E-8
    return (c (t + h) - c (t)) / h
```

### Tangenten (numerisch) interaktiv darstellen

```python
def InteractTangentNum (curve, a, b):
    InteractTangentSym (curve, lambda t: NumDiff (curve, t), a, b)
```

## Raumkurven

```python
# PlotVector3D

def PlotVector3D (p, q, **kwargs):
    q = np.hstack ((p, p + v))
    return pl.plot (q [0], q [1], q [2], **kwargs)
```

```python
# InteractTangent3D  (curve, a, b)

def InteractTangent3D (curve, a, b):

    tangent = lambda t: NumDiff (curve, t)
        
    tt = np.linspace (a, b, 1000)
    pp = curve (tt)
    qq = pp + tangent (tt)
    pq = np.hstack ((pp, qq))

    fig = Fig()
    ax = fig.add_subplot (111, projection = '3d')
    ax.set_box_aspect ((np.ptp (pq [0]), np.ptp (pq [1]), np.ptp (pq [2])))

    pl.plot (pp [0], pp [1], pp [2], color = 'blue', zorder = 2)
    pl.plot (qq [0], qq [1], qq [2], 'red', linewidth = 0)

    t0 = a
    p = curve (t0)
    v = tangent (t0)
    pnt, = pl.plot (p [0], p [1], p [2], 'o', color = 'red', zorder = 4) 
    tng, = pl.plot (p, v, color = 'red', zorder = 4)
    pl.show()

    def Callback (t0):

        p = curve (t0)
        v = tangent (t0)
        pv = np.hstack ((p, p + v))

        pnt.set_data_3d (p [0], p [1], p [2])
        tng.set_data_3d (pv [0], pv [1], pv [2])

    lay = iw.Layout (width = '900px')
    slt = iw.FloatSlider (min = a, max = b, step = (b - a) / 1000, value = (b - a) / 4, layout = lay)
    iw.interact (Callback, t0 = slt);
```

```python
def Helix (t):
    r = 1
    h = 1
    x = r * np.cos (2 * np.pi * t)
    y = r * np.sin (2 * np.pi * t)
    z = h * t
    return np.array ([[x], [y], [z]])

InteractTangent3D (lambda t: Helix (t / 10), 0, 20)
```
