```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
def f (x):
    return np.cosh (x)
```

```python
def f1 (x):
    return np.sinh (x)
```

```python
def f2 (x):
    return np.cosh (x)
```

```python
x = np.linspace (-2, 2, 1000)
```

```python
y = f (x)
```

```python
# hier ändern wir den betrachteten Punkt auf der Kurve
x0 = -0.5
```

```python
p = np.array ([
    [x0],
    [f (x0)]
])
p
```

```python
t = np.array ([
    [1],
    [f1 (x0)]
])
```

```python
n = np.array ([
    [- t [1] [0]],
    [  t [0] [0]]
])
```

```python
n = n / np.sqrt (n [0] [0] ** 2 + n [1] [0] ** 2)
```

```python
k = f2 (x0) / (1 + f1 (x0) ** 2) ** 1.5
```

```python
r = 1 / k
```

```python
m = p + r * n
```

```python
tt = np.linspace (0, 2 * np.pi, 1000)
```

```python
xx = m [0] [0] + r * np.cos (tt)
yy = m [1] [0] + r * np.sin (tt)
```

```python
pl.axis ('equal')
pl.plot (x, y)
pl.plot (p [0], p [1], 'o')
pl.plot ((p [0], p [0] + t [0]), (p [1], p [1] + t [1]))
pl.plot ((p [0], p [0] + n [0]), (p [1], p [1] + n [1]))
pl.plot ((m [0], p [0]), (m [1], p [1]))
pl.plot (xx, yy)
```
