```python
# Bibliotheken

import numpy as np
import matplotlib.pyplot as pl
import ipywidgets as iw

%matplotlib widget
```

```python
# figure

def Fig ():

    fig = pl.figure (figsize = (9, 5))

    cnv = fig.canvas
    cnv.toolbar_visible = False
    cnv.header_visible = False
    cnv.footer_visible = False

    return fig
```

## Beispiele ebener Kurven

```python
# CurveDemo

def CurveDemo (curve, a, b):

    tt = np.linspace (a, b, 200)
    pp = curve (tt)

    Fig()
    pl.axis ('equal')
    pl.plot (pp [0], pp [1], color = 'blue', zorder = 2)
    pnt, = pl.plot ([0], [0], 'o', color = 'red', zorder = 4) 

    def Callback (t0):
        p = curve (np.array ([t0]))
        pnt.set_data (p [0], p [1])

    lay = iw.Layout (width = '900px')
    sl = iw.FloatSlider (min = a, max = b, step = (b - a) / 1000, value = a, layout = lay)
    iw.interact (Callback, t0 = sl)
```

```python
def Beispiel (t):
    x = np.sqrt (t)
    y = t - t ** 2
    return np.array ([x, y])

CurveDemo (Beispiel, 0, 1)
```

```python
def InftySign1 (t):
    x = t * (t ** 2 - 49) * (21888 + 1532 * t ** 2 - 35 * t ** 4) / 2471040
    y = t * (t ** 2 - 49) * (t ** 2 - 16) * (108 - t ** 2) / 112320
    return np.array ([x, y])

CurveDemo (InftySign1, -7, 7)
```

```python
def InftySign2 (t):
    return np.array ([2 * np.sin (t), np.sin (2 * t)])
    
CurveDemo (InftySign2, 0, 2 * np.pi)
```

```python
def Neil (t):
    return np.array  ([t ** 3, 2 * t ** 2])

CurveDemo (Neil, -2, 2)
```

## Ableitung von Kurven, Tangentialvektor

```python
def Diff (c, t):
    h = 1E-8
    return (c (t + h) - c (t)) / h
```

```python
# PlotVector

def PlotVector (p, v, **kwargs):
    wd = 0.003
    return pl.quiver (p [0], p [1], v [0], v [1], **kwargs,
        angles = 'xy', scale = 1, scale_units = 'xy', width = wd, headwidth = 6, headlength = 10, headaxislength = 9)
```

```python
# TangentDemo

def TangentDemo (curve, a, b):

    tangent = lambda t: Diff (curve, t)

    tt = np.linspace (a, b, 200)
    pp = curve (tt)
    qq = pp + tangent (tt)

    Fig()
    pl.axis ('equal')
    pl.plot (pp [0], pp [1], color = 'blue', zorder = 2)
    pl.plot (qq [0], qq [1], 'red', linewidth = 0)

    t0 = a
    p = curve (t0)
    v = tangent (t0)
    pnt, = pl.plot (p [0], p [1], 'o', color = 'red', zorder = 4) 
    tng = PlotVector (p, v, color = 'red', zorder = 4)

    def Callback (t0):

        t = np.array ([t0])
        p = curve (t)
        v = tangent (t)

        pnt.set_data (p [0], p [1])
        tng.set_offsets ([p [0] [0], p [1] [0]])
        tng.set_UVC (v [0] [0], v [1] [0])

    lay = iw.Layout (width = '900px')
    sl = iw.FloatSlider (min = a, max = b, step = (b - a) / 1000, value = a, layout = lay)
    iw.interact (Callback, t0 = sl)
```

```python
TangentDemo (Beispiel, 0.05, 1)
```

```python
TangentDemo (lambda t: Beispiel (t / 10), 0.5, 10)
```

```python
TangentDemo (InftySign1, -7, 7)
```

```python
TangentDemo (InftySign2, 0, 2 * np.pi)
```

```python
TangentDemo (Neil, -2, 2)
```

## Raumkurven

```python
# PlotVector3D

def PlotVector3D (p, v, **kwargs):
    q = np.hstack ((p, p + v))
    return pl.plot (q [0], q [1], q [2], **kwargs)
```

```python
# TangentDemo3D

def TangentDemo3D (curve, a, b):

    tangent = lambda t: Diff (curve, t)
        
    tt = np.linspace (a, b, 1000)
    pp = curve (tt)
    qq = pp + tangent (tt)
    pq = np.hstack ((pp, qq))

    fig = Fig()
    ax = fig.add_subplot (111, projection = '3d')
    ax.set_box_aspect ((np.ptp (pq [0]), np.ptp (pq [1]), np.ptp (pq [2])))

    pl.plot (pp [0], pp [1], pp [2], color = 'blue', zorder = 2)
    pl.plot (qq [0], qq [1], qq [2], 'red', linewidth = 0)

    t0 = a
    p = curve (t0)
    v = tangent (t0)
    pnt, = pl.plot (p [0], p [1], p [2], 'o', color = 'red', zorder = 4) 
    tng, = PlotVector3D (p, v, color = 'red', zorder = 4)

    def Callback (t0):

        p = curve (t0)
        v = tangent (t0)
        pv = np.hstack ((p, p + v))

        pnt.set_data_3d (p [0], p [1], p [2])
        tng.set_data_3d (pv [0], pv [1], pv [2])

    lay = iw.Layout (width = '900px')
    slt = iw.FloatSlider (min = a, max = b, step = (b - a) / 1000, value = (b - a) / 4, layout = lay)
    iw.interact (Callback, t0 = slt);
```

```python
def Spirale (t):
    r = 1
    h = 1
    x = r * np.cos (2 * np.pi * t)
    y = r * np.sin (2 * np.pi * t)
    z = h * t
    return np.array ([[x], [y], [z]])

TangentDemo3D (lambda t: Spirale (t / 10), 0, 20)
```
