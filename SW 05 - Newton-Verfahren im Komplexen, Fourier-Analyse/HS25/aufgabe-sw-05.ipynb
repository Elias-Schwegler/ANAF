```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
def f (z):
    return 4 * z ** 3 - 2 * z ** 2 + z - 16
```

```python
def df (z):
    return 12 * z ** 2 - 4 * z + 1
```

```python
z0 = complex (-1, -1)
```

```python
def newton (z):
    eps = 1E-5

    z = z0
    #print (z)
    while True:
        dz = f (z) / df (z)
        z -= dz
        #print (z)
        if abs (dz) < 0.0001:
            break
    return z
```
