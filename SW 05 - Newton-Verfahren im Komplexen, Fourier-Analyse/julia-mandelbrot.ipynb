## Zu welcher komplexen Nullstelle konvergiert das Newton-Verfahren?

### Polynom

Nullstellen

$\qquad
Z_1 \;=\; 1 + i \,,
    \qquad Z_2 \;=\; -1 \,,
    \qquad Z_3 \;=\; 1 - i
$

Produkt der Linearfaktoren ausmultiplizieren

$\qquad\begin{align*}
    p \, (z)
        &\;=\; (z - Z_1) \cdot (z - Z_2) \cdot (z - Z_3) \\[6pt]
        &\;=\; (z - (1 + i)) \cdot (z - (-1)) \cdot (z - (1 - i))
        \;=\; (z - 1 - i) \cdot (z + 1) \cdot (z - 1 + i) \\[6pt]
        &\;=\; ((z - 1) - i) \cdot ((z - 1) + i) \cdot (z + 1)
        \;=\; ((z - 1) ^ 2  - i^2) \cdot (z + 1) \\[6pt]
        &\;=\; (z^2 - 2 \, z + 1 + 1) \cdot (z + 1)
        \;=\; (z^2 - 2 \, z + 2) \cdot (z + 1) \\[6pt]
        &\;=\; (z^3 - 2 \, z^2 + 2 \, z) + (z^2 - 2 \, z + 2) \\[6pt]
        &\;=\; z^3 - z^2 + 2 \\
\end{align*}$

Ableitung

$\qquad
    p' \, (z)
        \;=\; 3 \, z^2 - 2 \, z
$

### Newtonverfahren

#### Gleichung für die Iteration

$\qquad\displaystyle
    z_{n + 1}
        \;=\; z_{n} - \frac{p \, (z_{n})}{p' \, ({z_n})}
        \;=\; z_{n} - \frac{z_{n}^3 - z_{n}^2 + 2}{3 \, {z_n}^2 - 2 \, z_{n}}
$

siehe auch: https://de.wikipedia.org/wiki/Newtonverfahren

#### Implementation

```python
Z1 = complex ( 1,  1)
Z2 = complex (-1,  0)
Z3 = complex ( 1, -1)
```

```python
def Newton3 (z):                     # Startwert z
    eps = 1E-5                       # sehr kleine Zahl fuer den Vergleich mit 0
    for i in range (100):            # Iterationen
        f0 = z ** 3 - z ** 2 + 2     # Funktionswert an der Stelle z
        f1 = 3 * z ** 2 - 2 * z      # Ableitung an der Stelle z
        if abs (f1) < eps:           # verschwindet die Ableitung?
            break                    # falls zutreffend: Abbruch, um Division durch 0 zu vermeiden
        z -= f0 / f1                 # naechster Naeherungswert des Newton-Verfahrens
        if abs (z - Z1) < eps:       # Konvergenz zur ersten Nullstelle?
            return 1                 # Funktion mit Rueckgabewert 1 beenden
        if abs (z - Z2) < eps:       # Konvergenz zur zweiten Nullstelle?
            return 2                 # Funktion mit Rueckgabewert 2 beenden
        if abs (z - Z3) < eps:       # Konvergenz zur dritten Nullstelle?
            return 3                 # Funktion mit Rueckgabewert 3 beenden
    return 0                         # (bislang) keine Konvergenz, Funktion mit Rueckgabewert 0 beenden
```

### Test mit einzelnen Startwerten

```python
def Test1 (x0, y0):

    z0 = complex (x0, y0)
    r = Newton3 (z0)
    if r <= 0:
        print ('Keine Konvergenz')
    else:
        print ('Konvergenz zur Nullstelle {}, d1 = {:.2f}, d2 = {:.2f}, d3 = {:.2f}'
            .format (r, abs (z0 - Z1), abs (z0 - Z2), abs (z0 - Z3)))

    ctab = ['black', 'red', 'green', 'blue']
    pl.figure (figsize = (4, 3))
    pl.plot (x0, y0, 'o', color = ctab [r], fillstyle = 'none')
    pl.xlim ([-3.5, 3.5])
    pl.ylim ([-2.7, 2.7])
    pl.grid ('both')

    def p (x, y, c):
        pl.plot (x, y, 'o', color = ctab [c])
    p ( 1,  1,  1)
    p (-1,  0,  2)
    p ( 1, -1,  3)
```

#### Startwert $3 + 2 \, i$

```python
Test1 (3, 2)
```

Das Verfahren mit dem **Startwert** $z_0 = 3 + 2 \, i$ konvergiert zur **Nullstelle** $Z_1 = 1 + i$. <br>
Von allen Nullstellen hat $Z_1$ die **kleinste Entfernung** zum Startwert.

#### Startwert $i - 2$

```python
Test1 (-2, 1)
```

Das Verfahren mit dem **Startwert** $z_0 = i - 2$ konvergiert zur **Nullstelle** $Z_2 = -1$. <br>
Von allen Nullstellen hat $Z_2$ die **kleinste Entfernung** zum Startwert.

#### Startwert $2 - i$

```python
Test1 (2, -1)
```

Das Verfahren mit dem **Startwert** $z_0 = 2 - i$ konvergiert zur **Nullstelle** $Z_3 = 1 - i$. <br>
Von allen Nullstellen hat $Z_3$ die **kleinste Entfernung** zum Startwert.

#### Startwert $2 \, i - 1$

```python
Test1 (-1, 2)
```

**Überraschung**: Das Verfahren mit dem **Startwert** $z_0 = 2 \, i - 1$ konvergiert zur **Nullstelle** $Z_3 = 1 - i$. <br>
Von allen Nullstellen hat $Z_3$ nicht die geringste, sondern sogar die **grösste Entfernung** zum Startwert.

#### Startwert $0.6 \, i - 0.2$

```python
Test1 (-0.2, 0.6)
```

Das Verfahren mit dem **Startwert** $z_0 = 0.6 \, i - 0.2$ konvergiert zur **Nullstelle** $Z_1 = 1 + i$. <br>
Von allen Nullstellen hat $Z_1$ die **mittlere Entfernung** zum Startwert.

### Test mit vielen Startwerten

```python
def Test2 ():

    ctab = ['black', 'red', 'green', 'blue']

    def t (x0, y0):
        z0 = complex (x0, y0)
        r = Newton3 (z0)
        pl.plot (x, y, '.', color = ctab [r])
    pl.axis ('equal')
    pl.grid ('both')

    d = 0.2
    for x in np.arange (-4, 4 + d, d):
        for y in np.arange (-3, 3 + d, d):
            t (x, y)

    def p (x, y, c):
        pl.plot (x, y, 'o', color = ctab [c])
    p ( 1,  1,  1)
    p (-1,  0,  2)
    p ( 1, -1,  3)


Test2()
```

## Bilder zur Konvergenz des Newtonverfahrens

### Grauwertbilder

```python
def Plot1 (newton, xmin, xmax, ymin, ymax, delta):
    dx = xmax - xmin; n = int (dx / delta + 0.5); dx /= n; n += 1
    dy = ymax - ymin; m = int (dy / delta + 0.5); dy /= m; m += 1
    img = np.zeros ((m, n))
    for i in range (m):
        y = ymax - i * dy
        for k in range (n):
            x = xmin + k * dx
            z = complex (x, y)
            img [i, k] = newton (z)
    pl.imshow (img, cmap = 'grey')
    x1 = np.arange (np.ceil (xmin), np.floor (xmax) + 1, step = 1); x2 = (x1 - xmin) / dx; pl.xticks (x2, x1)
    y1 = np.arange (np.ceil (ymin), np.floor (ymax) + 1, step = 1); y2 = (ymax - y1) / dy; pl.yticks (y2, y1)

Plot1 (Newton3, -5, 5, -4, 4, 0.02)
```

```python
def Plot2 (newton, xmin, xmax, ymin, ymax, delta, colormap):
    dx = xmax - xmin; n = int (dx / delta + 0.5); dx /= n; n += 1
    dy = ymax - ymin; m = int (dy / delta + 0.5); dy /= m; m += 1
    img = np.zeros ((m, n, 3))
    for i in range (m):
        y = ymax - i * dy
        for k in range (n):
            x = xmin + k * dx
            z = complex (x, y)
            r = newton (z)
            img [i, k, :] = colormap [r]
    pl.imshow (img, cmap = 'grey')
    x1 = np.arange (np.ceil (xmin), np.floor (xmax) + 1, step = 1); x2 = (x1 - xmin) / dx; pl.xticks (x2, x1)
    y1 = np.arange (np.ceil (ymin), np.floor (ymax) + 1, step = 1); y2 = (ymax - y1) / dy; pl.yticks (y2, y1)

Plot2 (Newton3, -5, 5, -4, 4, 0.02, [Black, Red, Green, Blue])
```

## Verallgemeinerung

```python
def Plot (f0, f1, zeros, xmin, xmax, ymin, ymax, delta, colormap):

    def Newton (z):
        eps = 1E-5
        for i in range (100):
            denom = f1 (z)
            if abs (denom) < eps:
                break
            numer = f0 (z)
            z -= numer / denom
            for i in range (len (zeros)):
                if abs (z - zeros [i]) < eps:
                    return i + 1
        return 0

    dx = xmax - xmin; n = int (dx / delta + 0.5); dx /= n; n += 1
    dy = ymax - ymin; m = int (dy / delta + 0.5); dy /= m; m += 1
    img = np.zeros ((m, n, 3))
    for i in range (m):
        y = ymax - i * dy
        for k in range (n):
            x = xmin + k * dx
            z = complex (x, y)
            r = Newton (z)
            img [i, k, :] = colormap [r]
    pl.imshow (img, cmap = 'grey')
    x1 = np.arange (np.ceil (xmin), np.floor (xmax) + 1, step = 1); x2 = (x1 - xmin) / dx; pl.xticks (x2, x1)
    y1 = np.arange (np.ceil (ymin), np.floor (ymax) + 1, step = 1); y2 = (ymax - y1) / dy; pl.yticks (y2, y1)
```

## Einfachere Iteration: Mandelbrotmenge

### Iteration

$\qquad
    z_{n + 1}
        \;=\; z_{n}^2 + z_{0}
$

siehe auch: https://de.wikipedia.org/wiki/Mandelbrot-Menge

```python
def Mandelbrot (xmin, xmax, ymin, ymax, delta):

    def Iteration (z0):
        z = z0
        for i in range (100):
            z = z ** 2 + z0
            if abs (z) > 2:
                return 0
        return 1

    colormap = [Yellow, Black]

    dx = xmax - xmin; n = int (dx / delta + 0.5); dx /= n; n += 1
    dy = ymax - ymin; m = int (dy / delta + 0.5); dy /= m; m += 1
    img = np.zeros ((m, n, 3))
    for i in range (m):
        y = ymax - i * dy
        for k in range (n):
            x = xmin + k * dx
            z = complex (x, y)
            r = Iteration (z)
            img [i, k, :] = colormap [r]
    pl.imshow (img, cmap = 'grey')
    x1 = np.arange (np.ceil (xmin), np.floor (xmax) + 1, step = 1); x2 = (x1 - xmin) / dx; pl.xticks (x2, x1)
    y1 = np.arange (np.ceil (ymin), np.floor (ymax) + 1, step = 1); y2 = (ymax - y1) / dy; pl.yticks (y2, y1)

Mandelbrot (-2, 1, -1, 1, 0.01)
```
