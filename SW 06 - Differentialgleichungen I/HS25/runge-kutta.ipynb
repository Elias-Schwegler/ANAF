```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
def Ode (x, y):
    return x - y
```

```python
def Solution (c, x):
    return c / np.exp (x) + x - 1
```

```python
def Constant (x0, y0):
    return (y0 - x0 + 1) * np.exp (x0)
```

```python
X0 = 0
Y0 = 4
Xmax = 5

C = Constant (X0, Y0)
```

```python
def Euler (ode, x0, y0, h, xmax):
    n = int ((xmax - x0) / h + 0.5)
    xx = np.zeros (n + 1); xx [0] = x0
    yy = np.zeros (n + 1); yy [0] = y0
    for i in range (n):
        xx [i + 1] = xx [i] + h
        yy [i + 1] = yy [i] + h * ode (xx [i], yy [i])
    return [xx, yy]
```

```python
def Test (ode, sol, method):
    for k in range (5):
        h = 0.1 ** (k + 1)
        [xx, yy] = method (Ode, X0, Y0, h, Xmax)
        dd = sol (C, xx) - yy
        e = np.max (np.abs (dd))
        print ('stepsize = {stp:.2e}, error = {err:.2e}' .format (stp = h, err = e)) 
```

```python
Test (Ode, Solution, Euler)
```

```python
def RungeKutta (ode, x0, y0, h, xmax):
    n = int ((xmax - x0) / h + 0.5)
    xx = np.zeros (n + 1); xx [0] = x0
    yy = np.zeros (n + 1); yy [0] = y0
    h2 = h / 2
    for i in range (n):
        x0 = xx [i]
        x1 = x0 + h2
        x2 = x0 + h
        y0 = yy [i]
        k1 = ode (x0, y0)
        k2 = ode (x1, y0 + h2 * k1)
        k3 = ode (x1, y0 + h2 * k2)
        k4 = ode (x2, y0 + h * k3)
        xx [i + 1] = x2
        yy [i + 1] = y0 + h * (k1 + 2 * (k2 + k3) + k4) / 6
    return [xx, yy]
```

```python
Test (Ode, Solution, RungeKutta)
```
