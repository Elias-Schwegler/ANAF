# Triginometrische Funktionen

```python
#Bibliotheken

import numpy as np
import matplotlib.pyplot as pl
import matplotlib.ticker as tk
import ipywidgets as iw
%matplotlib widget
```

```python
def TrigFuncInteractive ():

    phi = 30
    res = np.pi / 180
    lw1 = 1.0
    lw2 = 2.0
    lw3 = 3.0

    def args (t0, t1):
        m = int (np.ceil (np.abs (t0 - t1) / res)) + 1
        n = ((m + 15) // 16) * 16 + 1
        return np.linspace (t0, t1, n, endpoint = True)
    
    def circle (r, t):
        x = r * np.cos (t)
        y = r * np.sin (t)
        return x, y

    def divide (nn, dd):
        ok = np.abs (dd) > 1E-10
        ee = np.full_like (dd, np.nan)
        return np.divide (nn, dd, out = ee, where = ok)

    pl.close()
    fig = pl.figure (figsize = (10, 4))
    cnv = fig.canvas
    cnv.toolbar_visible = False
    cnv.header_visible = False
    cnv.footer_visible = False

    axs1 = pl.subplot2grid ((1, 5), (0, 0), colspan = 2)
    axs1.axis ('equal')
    axs1.grid (which = 'both', axis = 'both', linestyle = ':')
    axs1.xaxis.set_major_locator (tk.MultipleLocator (1.0))
    axs1.xaxis.set_minor_locator (tk.MultipleLocator (0.5))
    axs1.yaxis.set_major_locator (tk.MultipleLocator (1.0))
    axs1.yaxis.set_minor_locator (tk.MultipleLocator (0.5))

    axs2 = pl.subplot2grid ((1, 5), (0, 2), colspan = 3)
    axs2.axis ('equal')
    axs2.grid (which = 'both', axis = 'both', linestyle = ':')
    axs2.xaxis.set_major_locator (tk.MultipleLocator (np.pi))
    axs2.xaxis.set_minor_locator (tk.MultipleLocator (np.pi / 4))
    axs2.yaxis.set_major_locator (tk.MultipleLocator (1.0))
    axs2.yaxis.set_minor_locator (tk.MultipleLocator (0.5))
    axs2.yaxis.tick_right()

    tt1 = args (0, 2 * np.pi)
    xx1, yy1 = circle (1, tt1)

    axs1.plot (xx1, yy1, 'gray', linewidth = lw1)
    xax1, = axs1.plot ([-2, 2], [0, 0], 'gray', linewidth = lw1)
    yax1, = axs1.plot ([0, 0], [-2, 2], 'gray', linewidth = lw1)
    ndl1, = axs1.plot ([], [], 'black', linewidth = lw1)
    arc1, = axs1.plot ([], [], 'black', linewidth = lw2)
    prj1, = axs1.plot ([], [], 'gray', linewidth = lw1)
    crs1, = axs1.plot ([], [], 'o', color = 'black', mfc = 'white', ms = 5)
    sin1, = axs1.plot ([], [], 'green', linewidth = lw2)
    cos1, = axs1.plot ([], [], 'red', linewidth = lw2)
    tan1, = axs1.plot ([], [], 'blue', linewidth = lw2)
    cot1, = axs1.plot ([], [], 'darkorange', linewidth = lw2)

    tt2 = args (-2 * np.pi, 2 * np.pi)
    xx2, yy2 = circle (1, tt2)
    zz1 = divide (yy2, xx2)
    zz2 = divide (xx2, yy2)

    axs2.plot ([tt2 [0], tt2 [-1]], [0, 0], 'gray', linewidth = lw1)
    xax2, = axs2.plot ([], [], 'gray', linewidth = lw1)
    yax2, = axs2.plot ([], [], 'gray', linewidth = lw1)
    ndl2, = axs2.plot ([], [], 'black', linewidth = lw1)
    arc2, = axs2.plot ([], [], 'black', linewidth = lw2)
    crs2, = axs2.plot ([], [], 'o', color = 'black', mfc = 'white', ms = 5)
    sin2, = axs2.plot ([], [], 'green')
    cos2, = axs2.plot ([], [], 'red')
    tan2, = axs2.plot ([], [], 'blue')
    cot2, = axs2.plot ([], [], 'darkorange')

    lay1 = iw.Layout (width = '700px')
    lay2 = layout = iw.Layout (width = '210px')
    lay3 = iw.Layout (width = '150px')
    
    sldPhi = iw.FloatSlider (description = 'continuous', layout = lay1, min = -360, max = 360, steps = 7200, value = phi)
    optPhi = [ '',
          '0°   (0)',       '30°   (1/6 pi)',  '45°   (1/4 pi)',  '60°   (1/3 pi)', 
         '90°   (1/2 pi)', '120°   (2/3 pi)', '135°   (3/4 pi)', '150°   (5/6 pi)', 
        '180°   (pi)',     '210°   (7/6 pi)', '225°   (5/4 pi)', '240°   (4/3 pi)', 
        '270°   (3/2 pi)', '300°   (5/3 pi)', '315°   (7/4 pi)', '330°   (11/6 pi)', 
        '360°   (2 pi)' ]
    ddnPhi = iw.Dropdown (description = 'discrete', layout = lay2, options = optPhi, value = optPhi [2])
    cbxNeg = iw.Checkbox (description = 'neg', layout = lay3, value = False)
    cbxSup = iw.Checkbox (description = 'sup', layout = lay3, value = False)
    cbxSin = iw.Checkbox (description = 'sin', layout = lay3, value = False)
    cbxCos = iw.Checkbox (description = 'cos', layout = lay3, value = False)
    cbxTan = iw.Checkbox (description = 'tan', layout = lay3, value = False)
    cbxCot = iw.Checkbox (description = 'cot', layout = lay3, value = False)

    def Continuous (phi, neg, sup, showSin, showCos, showTan, showCot):

        ddnPhi.disabled = True
        ddnPhi.value = ''
        ddnPhi.disabled = False
        
        if neg:
            phi = - phi
        if sup:
            phi = 180 - phi
        
        q0 = phi / 180
        t0 = q0 * np.pi
        x0, y0 = circle (1, t0)
        x1, y1 = circle (5, t0)
        print ('\n\tphi: %.2f° = %.5f pi = %.5f \t sin: %.5f, cos: %.5f, tan: %.5f, cot: %.5f'
            % (phi, q0, t0, y0, x0, divide (y0, x0), divide (x0, y0)))

        tt = args (0, t0)
        xx, yy = circle (1, tt)

        xax1.set_data ([-5, 5], [0, 0])
        yax1.set_data ([0, 0], [-5, 5])
        crs1.set_data ([x0], [y0])
        arc1.set_data (xx, yy)
        ndl1.set_data ([-x1, x1], [-y1, y1])
        
        xax2.set_data ([-7, 7], [0, 0])
        yax2.set_data ([0, 0], [-5, 5])
        crs2.set_data ([t0], [0])
        arc2.set_data ([0, t0], [0, 0])
        ndl2.set_data ([t0, t0], [-5, 5])

        if showSin:
            sin1.set_data ([x0, x0], [0, y0])
            sin2.set_data (tt2, yy2)
        else:
            sin1.set_data ([], [])
            sin2.set_data ([], [])

        if showCos:
            prj1.set_data ([x0, x0], [0, y0])
            cos1.set_data ([0, x0], [0,  0])
            cos2.set_data (tt2, xx2)
        else:
            prj1.set_data ([], [])
            cos1.set_data ([], [])
            cos2.set_data ([], [])

        if showTan:
            tan1.set_data ([1, 1], [0, divide (y0, x0)])
            tan2.set_data (tt2, zz1)
        else:
            tan1.set_data ([], [])
            tan2.set_data ([], [])

        if showCot:
            cot1.set_data ([0, divide (x0, y0)], [1, 1])
            cot2.set_data (tt2, zz2)
        else:
            cot1.set_data ([], [])
            cot2.set_data ([], [])

    def Discrete (val):

        if val == '':
            return
            
        str = val.split ('°') [0]
        sldPhi.value = float (str)

    ui = iw.VBox ([iw.HBox ([sldPhi, ddnPhi]), iw.HBox ([cbxNeg, cbxSup, cbxSin, cbxCos, cbxTan, cbxCot])])
    out = iw.interactive_output (Continuous, { 'phi': sldPhi, 'neg': cbxNeg, 'sup': cbxSup,
        'showSin': cbxSin, 'showCos': cbxCos, 'showTan': cbxTan, 'showCot': cbxCot })
    iw.interactive_output (Discrete, { 'val': ddnPhi,  })
    display (ui, out)

    pl.show()
```

```python
TrigFuncInteractive()
```
