# Hyperbelfunktionen vs. trigonometrische Funktionen

```python
import numpy as np
import matplotlib.pyplot as pl
```

## Der Kreis als Funktion

### Definition

#### Oberer Halbkreis

$ \displaystyle
    f_1 \, (x) \; = \; \sqrt{1 - x^2}
$

```python
def f1 (x):
    return np.sqrt (1 - x ** 2)
```

#### Unterer Halbkreis

$ \displaystyle
    f_2 \, (x) \; = \; - \, f_1 \, (x) \; = \; - \, \sqrt{1 - x^2}
$

```python
def f2 (x):
    return - f1 (x)
```

### Zeichnen

#### Wertetabelle einer Funktion

```python
def ValueTableFunction (f, a, b):
    n = 1000
    xx = np.linspace (a, b, n)
    yy = f (xx)
    return np.array ([
        xx, 
        yy
    ])
```

#### Wertetabelle zeichnen

```python
def PlotTable (table, col = 'black', lw = 2):
    pl.plot (table [0], table [1], color = col, linewidth = lw)
```

#### Funktion zeichnen

```python
def PlotFunction (f, a, b, col):
    PlotTable (ValueTableFunction (f, a, b), col)
```

#### Kreis aus zwei Halbkreisen zeichnen

```python
pl.plot ([-1.9, 1.9], [-1.4, 1.4], '.', color = 'white')   # Dummy fuer Achsenbegrenzungen
PlotFunction (f1, -1, 1, 'blue')
PlotFunction (f2, -1, 1, 'magenta')
pl.axis ('equal')
pl.grid ('both')
pl.show ()
```

## Erweiterung: Kreis mit Hyperbeln

### Definition

$ \displaystyle
    g_1 \, (x) \; = \; \sqrt{| \, 1 - x^2 \, |}
$

```python
def g1 (x):
    return np.sqrt (np.abs (1 - x ** 2))
```

$ \displaystyle
    g_2 \, (x) \; = \; - \, \sqrt{| \, 1 - x^2 \, |} \; = \; - \, g_1 (x)
$

```python
def g2 (x):
    return - g1 (x)
```

### Zeichnen

#### Linie zeichnen

```python
def PlotLine (x0, y0, x1, y1, col = 'black'):
    pl.plot ([x0, x1], [y0, y1], color = col)
```

#### Kreis mit Hyperbeln als Funktionen zeichnen

```python
PlotLine (-3, -3, 3,  3, 'gray')
PlotLine (-3,  3, 3, -3, 'gray')
PlotFunction (g1, -3, 3, 'blue')
PlotFunction (g2, -3, 3, 'magenta')
pl.axis ('equal')
pl.grid ('both')
pl.show ()
```

## Der Kreis als Kurve

### Definition

$ \displaystyle
    c \, (t) \; = \; \begin{pmatrix} \cos \, (t) \\ \sin \, (t) \end{pmatrix}
$

```python
def Circle (t):
    return np.array ([
        np.cos (t), 
        np.sin (t)
    ])
```

### Zeichnen

#### Wertetabelle einer Kurve

```python
def ValueTableCurve (c, a, b):
    n = 1000
    tt = np.linspace (a, b, n)
    return c (tt)
```

#### Kurve zeichnen

```python
def PlotCurve (c, a, b, col = 'black'):
    PlotTable (ValueTableCurve (c, a, b), col)
```

#### Kreis zeichnen

```python
pl.plot ([-1.9, 1.9], [-1.4, 1.4], '.', color = 'white')   # Dummy fuer Achsenbegrenzungen
PlotCurve (Circle, 0, 2 * np.pi, 'blue')
pl.axis ('equal')
pl.grid ('both')
pl.show ()
```

## Die Hyperbel als Kurve

### Definition

$ \displaystyle
    c \, (t) \; = \; \begin{pmatrix} \cosh \, (t) \\ \sinh \, (t) \end{pmatrix}
$

```python
def Hyperbola (t):
    return np.array ([
        np.cosh (t), 
        np.sinh (t)
    ])
```

### Kreis und Hyperbel als Kurven zeichnen

```python
PlotLine (0, 0, 3,  3, 'gray')
PlotLine (0, 0, 3, -3, 'gray')
PlotCurve (Circle, 0, 2 * np.pi, 'blue')
PlotCurve (Hyperbola, -1.8, 1.8, 'red')
pl.axis ('equal')
pl.grid ('both')
pl.show ()
```

## Vergleich mit der Kehrwertfunktion

### Funktion definieren

$\displaystyle
    h \, (x) \; = \; \frac{1}{x}
$

```python
def h (x):
    return 1 / x
```

### Zeichnen

#### Rotation definieren

$\displaystyle
    R \, (t) 
        \; = \;
        \begin{pmatrix*}[r] 
            \cos \, (t) & - \sin \, (t) \\
            \sin \, (t) &   \cos \, (t)
        \end{pmatrix*}
$

```python
def Rotate (phi):
    t = phi * np.pi / 180
    c = np.cos (t)
    s = np.sin (t)
    return np.array ([
        [c, -s], 
        [s,  c]
    ])
```

#### Skalierung definieren

$\displaystyle
    S \, (t) 
        \; = \;
        \begin{pmatrix*}[r] 
            t & 0 \\
            0 & t
        \end{pmatrix*}
$

```python
def Scale (f):
    return np.array ([
        [f, 0],
        [0, f]
    ])
```

#### Kehrwertfunktion und transformierte Hyperbel zeichnen

Hyperbel mit dem Faktor $\sqrt{2}$ skalieren und um $45^{\circ}$ bzw. $225^{\circ}$ drehen

```python
tab1a = ValueTableFunction (h, 0.2, 5)
tab1b = ValueTableFunction (h, -5, -0.2)
PlotTable (tab1a, 'gray', 5)
PlotTable (tab1b, 'gray', 5)

tab2 = Scale (np.sqrt (2)) @ ValueTableCurve (Hyperbola, -1.8, 1.8)
PlotTable (Rotate ( 45) @ tab2, 'red')
PlotTable (Rotate (225) @ tab2, 'red')

pl.axis ('equal')
pl.grid ('both')
pl.show ()
```
